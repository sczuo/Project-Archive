```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import re
import warnings

warnings.filterwarnings('ignore')
%pylab inline
```

```python
gun=pd.read_csv('texas.csv')
gun.describe()
sum(gun.isna())
gun.head()
```

```python
colnames=Series(gun.columns.values)
df_participant=gun[colnames[colnames.map(lambda x:x.startswith('part'))]]
df_participant['n_killed']=gun['n_killed']
df_participant['n_injured']=gun['n_injured']
df_participant.head()
```

```python
participant_type=gun['participant_type']
df_participant['n_victim']=participant_type.map(lambda x:len(re.findall('Victim',str(x))))
# check if n_victim=n_killed+n_injured
p=df_participant.copy()
len(p[p['n_victim']!=(p['n_killed']+p['n_injured'])]) #2740
df_participant['n_suspect']=participant_type.map(lambda x:len(re.findall('Subject-Suspect',str(x))))
```

```python
victim_gender=list(df_participant['participant_gender'].map(lambda x:re.findall('[a-zA-Z]+',str(x))))
for i in range(len(victim_gender)):
    j=df_participant['n_victim'][i]
    if j==0:
        victim_gender[i]=NaN
    else:
        victim_gender[i]=victim_gender[i][0:j]
victim_gender=Series(victim_gender)
df_participant['victim_gender']=victim_gender
```

```python
suspect_gender=list(df_participant['participant_gender'].map(lambda x:re.findall('[a-zA-Z]+',str(x))))
for i in range(len(victim_gender)):
    j=df_participant['n_suspect'][i]
    if j==0:
        suspect_gender[i]=NaN
    else:
        suspect_gender[i]=suspect_gender[i][-j:]
suspect_gender=Series(suspect_gender)
df_participant['suspect_gender']=suspect_gender
```

```python
age_group=df_participant['participant_age_group'].map(lambda x:re.findall('[a-zA-Z]+ \d+[-]*\d+[\+]*',str(x)))
len_each_age=age_group.map(lambda x:len(x))
idx=len_each_age[len_each_age==0].index.values
age_group[idx]=df_participant.iloc[idx]['participant_age_group']
vic_age_g={}
suspect_age_g={}
for i in range(len(age_group)):
    j=int(df_participant['n_suspect'][i])
    k=int(df_participant['n_victim'][i])
    if i not in idx:
        if j==0:
            suspect_age_g[i]=NaN
        else:
            suspect_age_g[i]=age_group[i][-j:]
        if k==0:
            vic_age_g[i]=NaN
        else:
            vic_age_g[i]=age_group[i][0:k]
suspect_age_g=Series(suspect_age_g)
vic_age_g=Series(vic_age_g)
df_participant['victim_age_group']=vic_age_g
df_participant['suspect_age_group']=suspect_age_g
```

```python
df_participant.iloc[0]
```

```python
df_participant['n_victim_male']=df_participant['victim_gender'].map(lambda x:len(re.findall('Male',str(x))))
df_participant['n_victim_female']=df_participant['victim_gender'].map(lambda x:len(re.findall('Female',str(x))))
df_participant['n_suspect_male']=df_participant['suspect_gender'].map(lambda x:len(re.findall('Male',str(x))))
df_participant['n_suspect_female']=df_participant['suspect_gender'].map(lambda x:len(re.findall('Female',str(x))))

df_participant['n_suspect_Child']=df_participant['suspect_age_group'].map(lambda x:len(re.findall('Child',str(x))))
df_participant['n_suspect_Teen']=df_participant['suspect_age_group'].map(lambda x:len(re.findall('Teen',str(x))))
df_participant['n_suspect_Adult']=df_participant['suspect_age_group'].map(lambda x:len(re.findall('Adult',str(x))))

df_participant['n_victim_Child']=df_participant['victim_age_group'].map(lambda x:len(re.findall('Child',str(x))))
df_participant['n_victim_Teen']=df_participant['victim_age_group'].map(lambda x:len(re.findall('Teen',str(x))))
df_participant['n_victim_Adult']=df_participant['victim_age_group'].map(lambda x:len(re.findall('Adult',str(x))))

```

```python
col=['n_victim','n_suspect','victim_gender','suspect_gender','victim_age_group','suspect_age_group',
     'n_victim_male','n_victim_female','n_suspect_male','n_suspect_female','n_suspect_Child','n_suspect_Teen',
    'n_suspect_Adult','n_victim_Child','n_victim_Teen','n_victim_Adult']
for c in col:
    gun[c]=df_participant[c]
```
